```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(X):
    return 4*X**2 - 5*X + 2

def derF(X):
    return 8*X-5
```

```python
x = np.linspace(-2,3,100)
```

```python
y = fn(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
X = -1.5
alpha = 0.24
px = [X]
py = [fn(X)]
```

```python
for i in range(40):
    X = X - alpha*derF(X)
    px.append(X)
    py.append(fn(X))
```

```python
plt.plot(x,y)
plt.plot(px,py,c='r')
plt.ylim(0,30)
plt.xlim(-3,4)
plt.show()
```

```python
w1 = 4
w0 = 2
alpha = 0.01
```

```python
x = np.array([2,3,6,7])
y = np.array([3,2,6,5])
```

```python
yp = w1*x+w0
```

```python
w0 = w0 - alpha*np.mean(yp-y)
```

```python
w1 = w1 - alpha*np.mean((yp-y)*x)
```

```python
yp = w1*x+w0
plt.scatter(x,y)
plt.plot(x,yp,c = 'r')
```
