````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

````python
print(sns.__version__)
````

````python
sns.get_dataset_names()
````

````python
tips_data = sns.load_dataset('tips')
````

````python
tips_data
````

````python
tips_data.groupby('day').mean(numeric_only=True)
````

````python
sns.barplot(tips_data,x = 'sex',y = 'tip',estimator='mean',hue='day',palette='crest')
````

````python
sns.barplot(tips_data,y = 'day',x = 'total_bill',estimator='mean',hue='time',palette='viridis',orient='h')
````

````python
sns.histplot(tips_data,y = 'total_bill',kde=True)
````

````python
sns.histplot(tips_data,x = 'tip',kde=True)
````

````python
sns.histplot(tips_data,x = 'tip',y='total_bill')
````

````python
sns.kdeplot(tips_data,x = 'total_bill')
````

````python
sns.kdeplot(tips_data,x = 'tip')
````

````python
sns.kdeplot(tips_data,x = 'tip',y='total_bill')
````

````python
sns.histplot(tips_data,x = 'total_bill',y = 'tip',color = 'red')
````

````python
sns.scatterplot(data = tips_data,x = 'total_bill',y = 'tip',hue='tip',legend=False)
````

````python
fmri_data = sns.load_dataset('fmri')
````

````python
fmri_data
````

````python
plt.figure(figsize=(10,10))
sns.lineplot(data = fmri_data,x = 'timepoint',y = 'signal',hue='event',style='region',markers=['*','o'])
````

````python
sns.jointplot(tips_data,x = 'total_bill',y = 'tip',kind='hex')
````

````python
sns.jointplot(tips_data,x = 'total_bill',y = 'tip',kind='scatter',hue='sex')
````

````python
sns.pairplot(tips_data,hue='day')
````

````python
sns.boxplot(tips_data,x = 'day',y = 'total_bill',hue='sex')
````

````python
sns.violinplot(tips_data,x = 'day',y = 'total_bill',hue='sex')
````

````python
sns.violinplot(tips_data,x = 'day',y = 'total_bill',split=True,hue = 'sex')
````

````python
sns.stripplot(tips_data,x = 'day',y = 'total_bill')
````

````python
sns.swarmplot(tips_data,x = 'day',y = 'total_bill')
````

````python
flights_data = sns.load_dataset('flights')
````

````python
flights_data
````

````python
nfd = flights_data.pivot(index = 'month',columns = 'year')
````

````python
nfd
````

````python
plt.figure(figsize=(10,10))
sns.heatmap(nfd,cmap = 'viridis',annot = True,fmt = 'g')
plt.show()
````

````python
sns.heatmap(tips_data.corr(numeric_only = True),annot = True)
````
