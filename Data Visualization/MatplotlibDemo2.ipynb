```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = np.linspace(-2,2,1000)
ySin= np.sin(2*np.pi*X)
yCos= np.cos(2*np.pi*X)

ytan= np.tan(2*np.pi*X)
ycot= yCos/ySin

plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.plot(X,ySin)
plt.grid()
plt.title("Sin Wave")

plt.subplot(2,2,2)
plt.plot(X,yCos)
plt.grid()
plt.title("Cos Wave")

plt.subplot(2,2,3)
plt.plot(X,ytan)
plt.grid()
plt.ylim((-5,5))
plt.title("Tan Wave")

plt.subplot(2,2,4)
plt.plot(X,ycot)
plt.grid()
plt.ylim((-5,5))
plt.title("Cot Wave")

plt.suptitle("Tignometric Functions")
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
X = np.linspace(-2,2,1000)
ySin= np.sin(2*np.pi*X)
yCos= np.cos(2*np.pi*X)

ytan= np.tan(2*np.pi*X)
ycot= yCos/ySin

fig,axes = plt.subplots(2,2,figsize=(12,6))
#fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(12,6))


axes[0,0].plot(X,ySin)
axes[0,0].grid()
axes[0,0].set_title("Sin Wave")

axes[0,1].plot(X,yCos)
axes[0,1].grid()
axes[0,1].set_title("Cos Wave")

axes[1,0].plot(X,ytan)
axes[1,0].grid()
axes[1,0].set_ylim((-5,5))
axes[1,0].set_title("Tan Wave")

axes[1,1].plot(X,ycot)
axes[1,1].grid()
axes[1,1].set_ylim((-5,5))
axes[1,1].set_title("Cot Wave")

plt.suptitle("Tignometric Functions")
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
X = np.array([10,20,30])
y2024 = np.array([80,60,100])
y2025 = np.array([75,80,90])

plt.bar(X-1.25,y2024,width=2.5,label = '2024')
plt.bar(X+1.25,y2025,width=2.5,label = '2025')

plt.xticks(X,['Gold', 'Silver', 'Platinum'])
plt.legend()
plt.show()
```

```python
X = np.array([10,20,30])
y2024 = np.array([80,60,100])
y2025 = np.array([75,80,90])

plt.barh(X-1.25,y2024,height=2.5,label = '2024')
plt.barh(X+1.25,y2025,height=2.5,label = '2025')

plt.yticks(X,['Gold', 'Silver', 'Platinum'])
plt.legend()
plt.show()
```

```python
X = [10,20,30,40,10]
plt.pie(X,labels = ['A','B','C','D','E'],autopct='%.2f%%',shadow=True,explode=[0,0,0,0,0.4])
plt.show()
```
